#### Step 1 : Create a tuple which contains the words as first parameter and the if the document is postive or negetive as the second parameter, and let's make a array of this.

```python
documents=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))

# Let's see first 5 records
documents[:5]
```

#### Step 2 : Shuffle the documents as currently they are 1000 neg followed by 1000 pos 

```python
import random
random.shuffle(documents)
documents[0:5]
```

#### Step 3 : Now we have to clean the data and lemmatize it

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
from nltk import pos_tag
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops, string.punctuation
```

#### Step 4 : We write a function to clean the data. It will take the above 2 functions which were created to remove stopwords and lemmatize

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words                                                    
```

#### Step 5 : Lemmatize and clean the data of all 2000 files

```python
documents = [(clean_review(document),category) for document, category in documents]
```

#### Step 6 : Now that we have cleaned the data, we will find feature words. That is we will first find all the words in all the files. Then for example we will take the top 3000 words(words with highest frequency). Then we will check if those 3000 words occour or not in each of the 2000 documents. This is how we will create a feature set.

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words = []
for doc in training_documents:
    all_words +=doc[0]
```

```python
import nltk
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
features
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
output = get_feature_dict(training_documents[0][0])
output
```

```python
training_data = [(get_feature_dict(doc),category) for doc, category in training_documents]
testing_data = [(get_feature_dict(doc),category) for doc, category in training_documents]
```

#### Step 7 : Now we have cleaned the data, we have found our features and thier frequency with each of the 2000 docs. Now we pass this data into a NaiveBayes Classifier and see how accurate the data is.

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```

```python
classifier.show_most_informative_features(15)
```

#### N-Grams

N-grams refer to bigrams,trigrams etc.
This means that when we were selecting top features, we were selecting most occouring 2000 words for example. When we use bigram we will select not only single words but also two continous words.
This will be extremely benifical in some cases like "not bad". Now if we had taken only single words "not" and "bad" would have been seperate and would not have had an impact, but now they do.

#### TF IDF

Refers to term-frequency and inverse term-frequency. These refer to frequency in each document rather than the overall frequency. Till now we were finding frequency for a word in all the documents. But this finds the frequency in each indivisual doument. For example if a word comes in each document it's frequency would be high but on the other hand we get a word which is there only in some documents but it's repeated multiple times within that document would be far more beneficial as this word can give us a very good idea if that particukar document is +ve or -ve because of that. So TF is benificial in that case.

Another advantage is that if we get one word which is repeating in all documents for eg "email" and its popping up in most occouring words, it's useless to us. so in that case we can use max_tf and min_tf. We can put max_tf =0.8. This means that if a word occours in>80% of the documents ignore it. IDF is just the inverse of tf. We multilply tf*idf so that even if some words escape in >80%. It's IDF will be very low if it's TF is high, which when multiplied will bring down the importnace of the word. SO this is also very usefull in somecases.
